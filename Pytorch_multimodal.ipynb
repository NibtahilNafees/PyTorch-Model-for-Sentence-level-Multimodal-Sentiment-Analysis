```python
import torch
import torchvision.models as models
import numpy as np

vgg_model = models.vgg16(pretrained=True)

from PIL import Image
import torchvision.transforms as transforms



# Define the transformation for the input image
transform = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
```

```python
dir="/content/drive/MyDrive/sentiment_images/"

def feature_extraction(filename):

    path = dir + filename
    # Open the image using PIL
    image = Image.open(path)
    if image.mode != 'RGB':
        image = image.convert('RGB')

    image = transform(image).unsqueeze(0)
#     image = torch.squeeze(image)

    
    vgg_model.eval()
    with torch.no_grad():
        features = vgg_model(image)

    features = features.squeeze().numpy()
    features = np.reshape(features, (features.shape[0], -1))
    
    print(f"Processed {filename}")

    return features
```

```python
def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct and not token.is_space]
    return tokens

def get_token_vectors(tokens):
    token_vectors = [nlp(token).vector for token in tokens]
    return token_vectors
```

```python
import torch.nn as nn
from transformers import BertModel, BertTokenizer

# load the BERT model and tokenizer
bert_model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def get_bert_embeddings(sentence):
    tokens = tokenizer.encode(sentence, add_special_tokens=True)

    # Convert the tokens to a PyTorch tensor
    input_ids = torch.tensor([tokens])

    with torch.no_grad():
        outputs = bert_model(input_ids)
        embeddings = outputs[0]

     # Take the mean of the embeddings tensor 
    mean_embeddings = torch.mean(embeddings, dim=((0,1)))

    return mean_embeddings
```

```python
#for train

text_embeddings_train = None
i = 0
for raw_text in train_data["raw"]:
    i = i + 1
    embeddings = get_bert_embeddings(raw_text)
    # Append the mean embeddings to the PyTorch tensor
    if text_embeddings_train is None:
        text_embeddings_train = embeddings
    else:
        text_embeddings_train = torch.vstack((text_embeddings_train, embeddings))
    
    # if i == 16:
    #     break

torch.save(text_embeddings_train,"text_embeddingstrain.pt")
```

```python
import torch.nn as nn

class ImageModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(1000, 512)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 256)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.5)
        self.fc3 = nn.Linear(256, 128)
        self.relu3 = nn.ReLU()

    def forward(self, images):
        x = self.fc1(images)
        x = self.relu1(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        x = self.relu3(x)
        return x
```

```python
class TextModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(768, 512)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 256)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.5)
        self.fc3 = nn.Linear(256, 128)
        self.relu3 = nn.ReLU()

    def forward(self, text):
        x = self.fc1(text)
        x = self.relu1(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        x = self.relu3(x)
        return x
```

```python
class Multimodal(nn.Module):
    def __init__(self,*args,**kwargs):
        super().__init__(*args,**kwargs)
        self.imageMod = ImageModel()
        self.textMod = TextModel()
        self.lastMod = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, text, images):
        m1 = self.textMod(text)
        m2 = self.imageMod(images)
        out = torch.cat([m1, m2], dim=1)
        out = self.lastMod(out)
        return out
```

```python
train_image_features = torch.load("train_imagetensor.pt")
test_image_features = torch.load("test_imagetensor.pt")
train_text_embeddings = torch.load("text_embeddingstrain.pt")
test_text_embeddings = torch.load("text_embeddingstest.pt")
train_labels = torch.load("trainlabels.pt")
test_labels = torch.load("testlabels.pt")
```

```python
train_image_features = train_image_features.reshape(len(train_labels),1000)
test_image_features = test_image_features.reshape(len(test_labels),1000)
# train_image_features.dtype
```

```python
print(train_image_features.shape , test_image_features.shape , train_text_embeddings.shape , test_text_embeddings.shape , train_labels.shape , test_labels.shape)
```

```python
train_loader=data_utils.DataLoader(data_utils.TensorDataset(train_image_features,train_text_embeddings,trainlabels),batch_size=4)
test_loader=data_utils.DataLoader(data_utils.TensorDataset(test_image_features,test_text_embeddings,testlabels),batch_size=4)
```

```python
mod= Multimodal()
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("GPU is available")
else:
    device = torch.device("cpu")
    print("GPU is not available")
mod.to(device)
```

```python
loss_func = nn.BCELoss()
learning_rate = 0.0001
optimizer = torch.optim.Adam(mod.parameters(), lr=learning_rate)
epochs = 1
```

```python
# Set the model to training mode
mod.train()
train_loss = []
accuracy_train = []
accuracy_test = []
for epoch in range(100):
 epoch_loss = 0.0
 for x1, x2, l in train_loader:
  x1, x2, l = x1.to(device), x2.to(device), l.to(device)
  optimizer.zero_grad()

  y_pred = mod(x2, x1)
  loss = loss_func(y_pred, l)
  loss.backward()
  
  
  optimizer.step()
  epoch_loss += loss.item()
 train_loss.append(epoch_loss / len(train_loader))
 mod.eval()
 correct_train, total_train = 0, 0
 correct_test, total_test = 0, 0

 for x1, x2, l in train_loader:
  x1, x2, l = x1.to(device), x2.to(device), l.to(device)
  outputs = mod(x2, x1)
  predicted = (outputs >= 0.5).float()
  total_train += l.size(0)
  
  correct_train += (predicted == l).sum().item()
 accuracy_train.append(correct_train / total_train)

 for x1, x2, l in test_loader:
  x1, x2, l = x1.to(device), x2.to(device), l.to(device)
  outputs = mod(x2, x1)
  
  predicted = (outputs >= 0.5).float()
  total_test += l.size(0)
  correct_test += (predicted == l).sum().item()
 accuracy_test.append(correct_test / total_test)

 print(f"Epoch {epoch+1}: Train Loss = {train_loss[-1]:.4f}, Train Acc = {accuracy_train[-1]*100:.2f}%, Test Acc = {accuracy_test[-1]*100:.2f}%")

#printing final loss
print('Last iteration loss value:', train_loss[-1])
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Set model to evaluation mode
mod.eval()

# Initialize variables for computing confusion matrix
true_labels = []
predicted_labels = []

# Make predictions on test set
with torch.no_grad():
    for x1, x2, l in test_loader:
        x1, x2, l = x1.to(device), x2.to(device), l.to(device)
        outputs = mod(x2, x1)
        predicted = (outputs >= 0.5).float()
        true_labels.extend(l.cpu().numpy())
        predicted_labels.extend(predicted.cpu().numpy())

# Compute confusion matrix and classification report
cm = confusion_matrix(true_labels, predicted_labels)
report = classification_report(true_labels, predicted_labels)

# Print confusion matrix, classification report, test accuracy, and F1 score
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(report)
test_acc = (cm[0][0] + cm[1][1]) / sum(sum(cm))
print("\nTest Accuracy:", test_acc)
f1 = 2 * cm[1][1] / (2 * cm[1][1] + cm[0][1] + cm[1][0])
print("F1 Score:", f1)
```
